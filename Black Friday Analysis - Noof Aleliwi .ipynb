# 2) Multi-Variate Analysis:

```python
    # DiScovering relationships between Gender and Purchases:
bf[['Gender','Purchase']].groupby(['Gender'], as_index = False).mean().sort_values(by='Purchase',ascending=False)
```

```python
    # DiScovering relationships between Material Status and Purchases:
bf[['Marital_Status','Purchase']].groupby(['Marital_Status'], as_index = False).mean().sort_values(by='Purchase',ascending=False)
```

```python
    # DiScovering relationships between City Category and Purchases:
bf[['City_Category','Purchase']].groupby(['City_Category'], as_index = False).mean().sort_values(by='Purchase',ascending=False)
```

Conclusions:

• Males Purachse more than women 
    
• Customers who live in city category C recorded the highest purchases
    
• The difference between single and married customers' buying power is unnoticable

# Problems in data quality:

```python
    # Deleting unneeded cols:
bf.drop(['User_ID','Occupation','Stay_In_Current_City_Years'], axis = 1, inplace=True)
```
